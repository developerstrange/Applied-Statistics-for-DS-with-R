```r
# Dataframe Remaining
df1 <- data.frame(
  ID = c(1,2,3,4),
  Age = c(22,33,23,22),
  Gender = c("F","F","M","M"),
  Location = c("rural","urban","rural","sub-urban"),
  stringsAsFactors = TRUE
)

df1
is.data.frame(df1)
summary(df1)

is.character(c("F","F","M","M"))
is.factor(as.factor(c("F","F","M","M")))      # converts character to factor(assign numeric value to each data) and checks is it a factor or not
is.numeric(as.factor(c("F","F","M","M")))    # checks if originally vector is a factor or converted to factor

# Subsetting
df1$ID
df1$Age
df1$Gender
df1$Location
mean(df1$Age)

df2 <- data.frame(
  ID = c(1,2,3,4),
  Age = c(22,33,23,NA),
  Gender = c("F","F","M","M"),
  Location = c("rural","urban","rural","sub-urban"),
  stringsAsFactors = TRUE
)
df2
mean(df2$Age, na.rm = TRUE)
```
