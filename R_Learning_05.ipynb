```r
math_operation <- function(x, y, op){
  # x = first argument
  # y = second argument
  # op = math operation name
  switch(
    op,
    "add" = x+y,
    "substract" = x-y,
    "multiply" = x*y,
    "division" = if(y!=0){x/y}else{warning("Division by zero!")},
    "Invalid operation"
  )
}

math_operation(2,3,"add")
math_operation(22,20,"Substraction")   # Case sensitive;
math_operation(33,0,"division")



cat("\n=====================================================\n")

# Setting default value to a functionn

calculate_sqrt <- function(val, digit = 2){
  # just set a default number to the argument which we want to be set by default.
  return (round(sqrt(val), digit))
  # rooted <- sqrt(val)
  # out <- round(rooted, digit)
  # return (out)
}

calculate_sqrt(1001)
cat("\n=====================================================\n")

# Scoping assignment
#   -used to declare a global variable inside a function

begin <- function(){
  begin_time <<- Sys.time()
  message("Program started: ",begin_time)
}

end <- function(){
  end_time <<- Sys.time()
  runtime <<- as.numeric(format(end_time, "%s")) - as.numeric(format(begin_time, "%s"))
  message("Program ended: ", end_time)
  message("Runtime stored ", runtime," in seconds")
}

begin()

```

```r
# Temporary Function

mat1 <- matrix(c(2,3,4,7), ncol = 2, nrow = 2, byrow = TRUE)

apply(mat1, 1, function(x) sum(x)^2)       # '1' indicates row
apply(mat1, 2, function(x) sum(x)^2)       # '2' indicates column


# Ellipsis
# When we need to pass multiple argument which will also be used in the function inside that function then we will ellipsis: '...'

wrapper_mean <- function(data, ...){
  if(is.numeric(data)){
    mean(data, ...)
  }else{
    warning("Data is not numeric")
  }
}

dt1 <- c(2,4,5,7,NA,99,55,7)

wrapper_mean(c(2,4,5,7,NA,99,55,7))
wrapper_mean(c(2,4,5,7,NA,99,55,'7'))
wrapper_mean(c(2,4,5,7,NA,99,55,'7'), na.rm = TRUE)
wrapper_mean(c(2,4,5,7,NA,99,55,7), na.rm = TRUE)
wrapper_mean(c(2,4,5,7,NA,99,55,7), na.rm = TRUE, trim = 0.4)


# Function as argument
# Function will be taken as argument and will be used inside the base function
apply_function <- function(vec,fun){
  return(fun(vec))
}

apply_function(c(2,4,5,7,22,99,55,7), mean)
```
