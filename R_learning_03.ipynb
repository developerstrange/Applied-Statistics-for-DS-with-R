```r
# Data Structures in R

#
# In R, based on data type the data structure can be divided into 2 categories:
#   1. Single data type (Vector, Matrix, Array)
#   2. Multiple data type (DataFrame, List)
```

```r
# Vector

x1 <- c(5,3,76,33,77,98,22)
is.vector(x1)
is.data.frame(x1)
```

```r
# Random number generator

set.seed(769)  # set.seed() is actually used for setting a static value so that others get the same random sample value using the same seed. if we don't use set.seed() then other's will get different different values every time.
x2 <- sample(1:100, 10)
x2
summary(x2)

x3 <- sample(1:10, 8, replace = TRUE)  # replace = True means same number can repeat multiple times
table(x3)  # used for count the numbers


x4 <- runif(5)  # used to generate random numbers between 0 and 1.
x4

prob_weights = c(4, 6, 2, 6, 1)
fruits <- sample(c("Apple", "Banana", "Mango", "Litchi", "Orange"), 3, replace = TRUE, prob = prob_weights)
fruits


```

```r
# Sequence Generator
x5 <- 1:15
x5
x6 <- 10:1
x6

x7 = seq(from = 1, to = 10, by = 3)  # 'by' defines the gap between numbers
x7

x8 = seq(from = 1, to = 15, length.out = 5) # 'length.out' defines a definite gap between numbers. we don't define it. it will be defined automatically
x8
```

```r
# Selection
x1 > 50
x1[c(FALSE, FALSE,TRUE, FALSE, TRUE, TRUE, TRUE)]
x1[c(2, 4, 1)]
x1[x1>50]
x1[x1> 30 & x1<50]
which(x1>30 & x1<50)
x1[which(x1>30 & x1<50)]
```

```r
# Replace

# Don't replace original
replace(x1, x1>30, 50)
x1 <- replace(x1, which(x1>30), NA)
x1

avg = mean(x1, na.rm = TRUE)
avg
# Replace original
x1[is.na(x1)] <- 10
x1
x9 <- c(5,3,76,33,77,98,22)
x9[x9 > 70] <- 58
x9
```

```r
# Matrix

mat1 <- matrix(c(3,2,4,1,8,4), ncol=3, nrow=2, byrow=TRUE)
print(mat1)
is.matrix(mat1)

is.matrix(c(0,2,3))

print(mat1[,])
print(mat1[2,])
print(mat1[,1])
print(mat1[2,2])
mat1[2,2] <- 0
print(mat1)

mat2 <- matrix(c(0,1,2), ncol = 1)
print(mat2)
```

```r
mat4 <- matrix(c(1,2,3,4,5,6), 2, 3, TRUE)
mat5 <- matrix(c(4,5,6,7,8,9), 3, 2, TRUE)
mat6 <- matrix(c(1,2,3),3, byrow = TRUE)
dim(mat5)
dim(mat6)
# Element wise multiplication(Hadamard Multiplicaiton)
mat7 <- matrix(c(2,2,3,4),2)
mat8 <- matrix(c(4,4,5,5),2)
print(mat7 * mat8)

# Matrix Multiplication
print(mat4 %*% mat5)

# Kronecker Product
print(mat4 %x% mat6)
```

```r
# Array

arr1 <- array(c(1:15), c(2,3,3))  # second c() funciton cointain c(row,column,layer)
print(arr1)
```

```r
# Data Frame

df1 = data.frame(
  ID = c(1,2,3),
  Age = c(24,23,26),
  Gender = c("M", "M", "F"),
  stringsAsFactors = TRUE
)
df1
summary(df1)

```
